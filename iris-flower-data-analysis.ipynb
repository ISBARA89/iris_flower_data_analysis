```python
import seaborn as sns
import pandas as pd

# Load the dataset
df = sns.load_dataset('iris')
```

```python
# Display the first few rows
print(df.head())

# Get summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Check the shape of the data after removing duplicates
print(df.shape)
```

```python
# Group by species and calculate mean
grouped_data = df.groupby('species').mean()
print(grouped_data)
```

```python
import matplotlib.pyplot as plt

# Histogram for each feature
df.hist(edgecolor='black', linewidth=1.2, figsize=(12,8))
plt.show()

# Boxplot to see the distribution by species
plt.figure(figsize=(12,8))
sns.boxplot(x='species', y='petal_length', data=df)
plt.show()
```
